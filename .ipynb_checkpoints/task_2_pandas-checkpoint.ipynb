1. Скачайте Titanic dataset и загрузите его в pandas.DataFrame; выведите первые 5 записей в ноутбук. Как можно интерпретировать данную таблицу?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from catboost.datasets import titanic #catboost onlt for dataset
```

```python
titanic_train, titanic_test = titanic()
df_train = titanic_train
df_test = titanic_test

display(df_train.head())
display(df_test.head())
```

Набор данных включает в себя два файла. Файл df_train — обучающий набор данных, файл df_test — тестовый набор данных. Обучающий набор содержит признак Survived для каждого пассажира, обозначающий, выжил данный пассажир или нет (0 для умерших, 1 для выживших).

Каждая строчка наборов данных содержит следующие поля:
Pclass — класс пассажира (1 — высший, 2 — средний, 3 — низший);
Name — имя;
Sex — пол;
Age — возраст;
SibSp — количество братьев, сестер, сводных братьев, сводных сестер, супругов на борту титаника;
Parch — количество родителей, детей (в том числе приемных) на борту титаника;
Ticket — номер билета;
Fare — плата за проезд;
Cabin — каюта;
Embarked — порт посадки (C — Шербур; Q — Квинстаун; S — Саутгемптон).
В поле Age приводится количество полных лет. Для детей меньше 1 года — дробное. Если возраст не известен точно, то указано примерное значение в формате xx.5.

10. Определите средний возраст пассажиров, медианный возраст, дисперсию.

```python
print(f"mean age оf passangers: {round(df['Age'].mean(),3)}")
print(f"median age оf passangers: {df['Age'].median()}")
print(f"variance in the age of passengers: {round(df['Age'].var(), 3)}")
```

11. Добавьте признак, который показывает, какой процент пассажиров имели возраст строго меньше, чем данный пассажир.

```python
df['Is_older'] = (df[df['Age'] > df['Age'].mean()]).shape[0] / df.shape[0]
df.head()
```

12. Коррелируют ли число братьев/сестер с числом родителей/детей?

```python
print('Коррелируют ли число братьев/сестер/супругов с числом родителей/детей')
sisp = df['SibSp']
parch = df['Parch']
corr = sisp.corr(parch, method='pearson')
print(f'корреляция: {round(corr, 2)}')
```

тк коэффициент корреляции меньше 0.5 можно предположить существует слабая связь между числом 
братьев/сестер/супругов с числом родителей/детей

14. Есть ли зависимость между классом и номером билета?

```python
df['Ticket'] = pd.to_numeric(df['Ticket'].str.replace('\D', ''), errors='coerce') 
tickets = df['Ticket']
pclasses = df['Pclass']
corr = tickets.corr(pclasses, method='pearson')
print(round(corr, 2))
```

15. Какой части пассажиров удалось выжить?

```python
survived = df[df.Survived==1]['PassengerId'].count()
dead =df[df.Survived==0]['PassengerId'].count()
print(f'Доля выживших пассажиров: {round(survived/(survived + dead), 2)}')
```

16. Сделайте визуализацию, позволяющую ответить на вопросы:
  * Верно ли, что женщины выживали чаще мужчин?


```python
import seaborn as sns
sns.catplot(data = df,hue = 'Survived', x = 'Sex', kind='count')
```

Верно, что женщины выживали чаще мужчин

  * Верно ли, что чаще выживали пассажиры с более дорогими билетами?

```python
sns.catplot(data = df, y='Survived', x='Sex', col='Pclass',kind='bar')
```
